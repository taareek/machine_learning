## Visualization 

```python
sns.pairplot(iris, hue = 'species', size=1.5)
```

```python
sns.pairplot(iris, hue = 'species', size=3, aspect= 1)
```

```python
sns.pairplot(iris, hue = 'species', diag_kind= 'hist', size=3, aspect= 1)
```

```python
# plotting the histogram for iris dataset
iris.hist(edgecolor= 'orange', linewidth= 1.2, figsize= (10,6));
```

```python
#iris.hist?
```

```python
# Violin plot

plt.figure(figsize= (12, 8))  # declaring the figure size
plt.subplot(2,2,1)
sns.violinplot(x = 'species',y= 'sepal_length', data= iris)
plt.subplot(2,2,2)
sns.violinplot(x= 'species', y= 'sepal_width', data= iris)
plt.subplot(2,2,3)
sns.violinplot(x= 'species', y= 'petal_length', data= iris)
plt.subplot(2,2,4)
sns.violinplot(x= 'species', y= 'petal_width', data= iris)
```

```python
# Box plot
iris.boxplot(by= 'species', figsize= (12,8));
```

```python
pd.plotting.scatter_matrix(iris, figsize=(12,8))
plt.show()
```
