Description of columns in our Bonston dataset

    1. CRIM      per capita crime rate by town
    2. ZN        proportion of residential land zoned for lots over 
                 25,000 sq.ft.
    3. INDUS     proportion of non-retail business acres per town
    4. CHAS      Charles River dummy variable (= 1 if tract bounds 
                 river; 0 otherwise)
    5. NOX       nitric oxides concentration (parts per 10 million)
    6. RM        average number of rooms per dwelling
    7. AGE       proportion of owner-occupied units built prior to 1940
    8. DIS       weighted distances to five Boston employment centres
    9. RAD       index of accessibility to radial highways
    10. TAX      full-value property-tax rate per 10,000(USD)
    11. PTRATIO  pupil-teacher ratio by town
    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
                 by town
    13. LSTAT    % lower status of the population
    14. MEDV     Median value of owner-occupied homes in 1000(USD)'s

# RANSAC (Random Sample Consensus Algorithm)

```python
X = data['RM'].values.reshape(-1, 1)
y = data['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X, y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3,10,1)
line_y_ransac = ransac.predict(line_X.reshape(-1, 1))
```

```python
sns.set(style= 'darkgrid', context= 'notebook')
plt.figure(figsize=(12,10))
plt.scatter(X[inlier_mask], y[inlier_mask], c= 'blue', marker= 'o', label= 'Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask], c= 'brown', marker= 's', label= 'Outliers')
plt.plot(line_X, line_y_ransac, color= 'red')
plt.xlabel('Average number of room per building')
plt.ylabel("Median value of owner occupied home in $1000's")
plt.legend(loc= 'upper left')
plt.show()
```

```python
# getting coefficient 
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

# Performance Evaluation of Regression Model

```python
from sklearn.model_selection import train_test_split
```

```python
X = data['LSTAT'].values.reshape(-1,1)
```

```python
for i in range(10):
    print(X[i])
```

```python
y = data['MEDV'].values
```

```python
for i in range(5):
    print(y[i])
```

```python
# splitting data into 80% train set and 20% test set
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print("Length of train set: ",len(X_train))
print("Length of test set: ",len(X_test))
```

```python
Y_test
```

```python
# ddefining Linear Regression
lr = LinearRegression()
```

```python
# train the model 
lr.fit(X_train, Y_train)
```

```python
# predict train set
y_train_predict = lr.predict(X_train)

# predict test set 
y_test_predict = lr.predict(X_test)
```

```python
y_test_predict
```

Method 1: Residual Analysis

```python
plt.figure(figsize= (12,8))
plt.scatter(y_train_predict, y_train_predict- Y_train, c='blue', marker='o', label= "Training Data")
plt.scatter(y_test_predict, y_test_predict- Y_test, c='red', marker='*', label= "Test Data")
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.legend(loc="upper left")
plt.hlines(y=0, xmin=-10, xmax= 50, lw= 2, color= 'k')
plt.xlim([-10, 50])
plt.show()
```

Method 2: Mean Square Error


```python
from sklearn.metrics import mean_squared_error
```

```python
# MSE for train data
train_mse = mean_squared_error(Y_train, y_train_predict)
print("MSE for train data: ", train_mse)
```

```python
# MSE for test data
test_mse = mean_squared_error(Y_test, y_test_predict)
print("MSE for test data: ", test_mse)
```

Coefficient of Determination, R^2

```python
from sklearn.metrics import r2_score
```

```python
# R2 score for train set
# the higher r2 score is better model (between 0-1)
r2_train = r2_score(Y_train, y_train_predict)
```

```python
print("R2 Score for train data: ", r2_train)
```

```python
r2_test = r2_score(Y_test, y_test_predict)
print("R2 Score for test data: ", r2_test)
```
