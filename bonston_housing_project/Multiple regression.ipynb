```python
# Importing necessary libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
# getting boston dataset 
from sklearn.datasets import load_boston
```

```python
boston_data = load_boston()
```

```python
# making dataframe for boston data
df = pd.DataFrame(boston_data.data, columns= boston_data.feature_names)
```

```python
# printing head of boston dataset from our dataframe 
df.head()
```

```python
# getting shape of boston dataset
df.shape
```

```python
# taking X as our features 
X = df
# taking y as our target 
y = boston_data.target
```

```python
# visualizing X=> which is our features
X
```

```python
print(len(X))
```

```python
# visualizing our trget variables for our features 
print(y)
```

### Statsmodels  

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

We need to add a constant term to allow statsmodels.api to calculate the bias/ intercepts

```python
# adding a constant 
X_constant = sm.add_constant(X)
```

```python
# visualizing features with constant
pd.DataFrame(X_constant)
```

```python
sm.OLS?
```

```python
# creating Ordinary Least Square Model(OLS)
model = sm.OLS(y, X_constant)
```

```python
# fit our model 
lr = model.fit()
```

```python
# printing the summary of our model 
lr.summary()
```

#### Using statsmodels.formula.api 

```python
# applying formula 
form_lr = smf.ols(formula= 'y~ CRIM+ ZN+ INDUS+ CHAS+ NOX+ RM+ AGE+ DIS+ RAD+ TAX+ PTRATIO+ B+ LSTAT', 
                 data= df)

# fit the model 
mlr = form_lr.fit()
```

```python
# getting summary
mlr.summary()
```

Now we will be taking some columns from our dataset and apply muliple Regression 

```python
cust_lr = smf.ols(formula= 'y~ CRIM+ ZN+ INDUS+ CHAS+ NOX+ RM',
                 data= df)

# fitting the model 
c_mlr= cust_lr.fit()
```

```python
# getting the summary
c_mlr.summary()
```

## Correlation Matrix

### Detecting Collinearity with Eigenvectors

```python
eigenvalues, eigenvectors= np.linalg.eig(df.corr())
```

```python
# printing eigenvalues using pandas  
pd.Series(eigenvalues).sort_values()
```

```python
# print eigen vectors 
np.abs(pd.Series(eigenvectors[:,8])).sort_values(ascending= False)
```

#### Standarize varibale to identify key features

In order to perform impact of the variable/ factor on the model properly, one needs to standarize the variable properly

```python
from sklearn.linear_model import LinearRegression 
```

```python
# defining model 
model = LinearRegression()
```

```python
# fit the model for our features
model.fit(X,y)
```

```python
result = pd.DataFrame(list(zip(model.coef_, df.columns)), columns= ['coeffcient', 'name']).set_index('name')
```

```python
result
```

```python
# printing coeffficeint as descending order
np.abs(result).sort_values(by= 'coeffcient', ascending= False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()
standard_coefficient_linear_reg = make_pipeline(scaler, model)
```

```python
# fit the model 
standard_coefficient_linear_reg.fit(X,y)
result = pd.DataFrame(list(zip(standard_coefficient_linear_reg.steps[1][1].coef_, df.columns)),
                     columns= ['coeffcient', 'name']).set_index('name')
np.abs(result).sort_values(by= 'coeffcient', ascending= False)
```

### Use $R^2$ to identify key features

```python
from sklearn.metrics import r2_score
```

```python
linear_reg = smf.ols(formula= 'y~ CRIM+ ZN+ INDUS+ CHAS+ NOX+ RM+ AGE+ DIS+ RAD+ TAX+ PTRATIO+ B+ LSTAT', 
                 data= df)

benchmark = linear_reg.fit()
r2 = r2_score(y, benchmark.predict(df))
print("R2_square value: ", r2)
```
